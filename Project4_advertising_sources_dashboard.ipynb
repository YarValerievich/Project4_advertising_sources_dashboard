# Анализ рекламных источников

В датасете содержатся данные первых пользователей приложения — когорты пользователей, которые начали пользоваться приложением в период с 4 по 10 мая включительно.

Проанализируем поведения игроков в зависимости от источника перехода.

- Проведем исследовательский анализ данных;
- Посчитаем количество построек для каждого пользователя;
- Проанализируем влияние источника перехода в игру на поведение пользователя;
- Сравним активность и стоимость привлечения пользователей с разбивкой по источникам трафика;
- Приоритизируем каналы по выручке для того, чтобы понять, в какие стоит вкладывать бюджет;
- Оценим, какие пользователи строят больше построек:
    - *завершившие уровень победой над врагом;*
    - *разработавшие орбитальную сборку спутников.*

Проверим статистические гипотезы.

1.  Время завершения уровня различается в зависимости от способа прохождения:

        H0: время завершения уровня через реализацию проекта = время через победу над первым игроком;
        H1: время завершения уровня через реализацию проекта ≠ время через победу над первым игроком.
    
    
2. Наличие взаимосвязи между стратегией развития у пользователей, прошедших уровень, и количеством построек:

        H0: взаимосвязь между стратегией развития и количеством построек отсутствует;
        H1: между стратегией развития и количеством построек взаимосвязь существует.
        
Подготовим промежуточные и итоговый выводы, оформим важные аспекты в виде презентации и создадим дашборд.

### Стоимость привлечения пользователей с разбивкой по источникам трафика

Подсчитаем количество новых пользователей за каждый день с 4 по 10 мая.

```python
tmp04 = df.query('day >= "2020-05-04" and day < "2020-05-05"').groupby(['source', 'day'])\
    .agg({'user_id': 'nunique'}).reset_index()
tmp04['count'] = tmp04['user_id'] 

tmp05 = df.query('day >= "2020-05-05" and day < "2020-05-06"').groupby(['source', 'day'])\
    .agg({'user_id': 'nunique'}).reset_index()
tmp05['count'] = tmp05['user_id'] - tmp04['user_id'] 

tmp06 = df.query('day >= "2020-05-06" and day < "2020-05-07"').groupby(['source', 'day'])\
    .agg({'user_id': 'nunique'}).reset_index()
tmp06['count'] = tmp06['user_id'] - tmp05['user_id'] 

tmp07 = df.query('day >= "2020-05-07" and day < "2020-05-08"').groupby(['source', 'day'])\
    .agg({'user_id': 'nunique'}).reset_index()
tmp07['count'] = tmp07['user_id'] - tmp06['user_id'] 

tmp08 = df.query('day >= "2020-05-08" and day < "2020-05-09"').groupby(['source', 'day'])\
    .agg({'user_id': 'nunique'}).reset_index()
tmp08['count'] = tmp08['user_id'] - tmp07['user_id'] 

tmp09 = df.query('day >= "2020-05-09" and day < "2020-05-10"').groupby(['source', 'day'])\
    .agg({'user_id': 'nunique'}).reset_index()
tmp09['count'] = tmp09['user_id'] - tmp08['user_id'] 

tmp10 = df.query('day >= "2020-05-10" and day < "2020-05-11"').groupby(['source', 'day'])\
    .agg({'user_id': 'nunique'}).reset_index()
tmp10['count'] = tmp10['user_id'] - tmp09['user_id'] 

tmp = pd.concat([tmp04, tmp05, tmp06, tmp07, tmp08, tmp09, tmp10])
tmp = tmp.sort_values(by=['source', 'day']).reset_index()
tmp = tmp[['source', 'day', 'count']]
```

```python
df_source_costs = tmp.merge(ad_costs['cost'],  left_index=True, right_index=True, how = 'left')
df_source_costs
```

Построим итоговую таблицу с ценой привлечения за каждого пользователя в разрезе источника и даты привлечения.

```python
df_source_costs['costs_for_user'] = df_source_costs['cost']/df_source_costs['count']
df_source_costs.sort_values(by='costs_for_user')
df_source_costs.query('source == "yandex_direct"')
```

```python
display(df_source_costs.groupby('source').agg({'costs_for_user': 'sum'}).reset_index()\
        .sort_values(by='costs_for_user', ascending=False)\
    .style.format({'costs_for_user': '{:.2f}'}))

df_source_costs.groupby('source')['costs_for_user'].sum()\
    .plot(kind='bar', color='#EAE77B', figsize=(10,6), grid=True, ylim=(6.5,13.5))

plt.xticks(rotation = 0)
plt.title('Средняя цена привлечения с разбивкой по источникам')
plt.xlabel('Источник привлечения')
plt.ylabel('Стоимость одного пользователя')
plt.show()  
```

Самым дорогим каналом привлечения является facebook_ads с ценой 13.23 евро за пользователя. После него идет instagram_new_adverts - 11.66 евро и yandex_direct - 7.32, а самым дешевым каналом оказался youtube_channel_reklama с ценой привлечения 6.97 евро.

```python
df_source_costs.groupby('day')['costs_for_user'].sum()\
    .plot(kind='bar', color='#EAE77B', figsize=(10,6), grid=True)

plt.xticks(rotation = -30)
plt.title('Средняя цена привлечения с разбивкой по датам')
plt.xlabel('Дата привлечения')
plt.ylabel('Стоимость одного пользователя')
plt.show()  

df_source_costs.groupby('day').agg({'costs_for_user': 'sum'}).reset_index().sort_values(by='costs_for_user', ascending=False)\
    .style.format({'day': '{:%Y-%m-%d}', 'costs_for_user': '{:.2f}'})
```

Наблюдаем, что со второго дня стоимость одного пользователя падает. Разобьём график на четыре по каналам привлечения.

```python
for i in df['source'].unique():
    
    print('Канал привлечения', i)
    
    df_source_costs.query('source == @i').groupby('day')['costs_for_user'].sum()\
        .plot(kind='bar', color='#EAE77B', figsize=(10,6), grid=True)

    plt.xticks(rotation = -30)
    plt.title('Средняя цена привлечения с разбивкой по датам')
    plt.xlabel('Дата привлечения')
    plt.ylabel('Стоимость одного пользователя')
    plt.show()  
    
    print()
```

Мы видим, что ситуация со снижением цены привлечения стабильна для всех источников привлечения за исключением `facebook_ads`.

### Приоритизация источников привлечения

Среднее количество построек на пользователя в разрезе по стратегии игрока и источник привлечения.

```python
df_source_building.groupby(['finish', 'source']).agg({'building_count':'mean'}).reset_index()\
    .sort_values(by='building_count', ascending=False)
```

Количество построек на пользователя с разбивкой по источникам привлечения практически не различаются.

Сумма построек пользователей в разрезе по стратегии игрока и источнику привлечения, а также количество пользователей, использующих ту или иную стратегию. 

```python
df_source_strategy = df_source_building.groupby(['finish', 'source'])\
    .agg({'building_count':'sum', 'user_id':'nunique'}).reset_index()
df_source_strategy.rename(columns={'user_id': 'users_count'}, inplace=True)
df_source_strategy
```

Посчитаем сумму игроков различных стратегий и найдем процентное соотношение количества игроков каждой стратегии из каждого источника.

```python
for i in range(1, 4):
    df_source_strategy.loc[df_source_strategy['finish'] == i, '%_finish_strategy'] = \
    round(df_source_strategy['users_count'] * 100 / df_source_strategy.query('finish == @i')['users_count'].sum() , 1)
    
df_source_strategy
```

## Статистические гипотезы

Уровень статистической значимости равен 0.05.

### Время завершения уровня различается в зависимости от способа прохождения

    H0: время завершения уровня через реализацию проекта = время через победу над первым игроком;
    H1: время завершения уровня через реализацию проекта ≠ время через победу над первым игроком.

Соберём таблицу, сгруппированную по пользователям, и посчитаем время завершения уровня.

    Стратегия 1 - пользователи, не закончившие уровень.
    Стратегия 2 - пользователи, завершившие уровень победой над врагом.
    Стратегия 3 - пользователи, завершившие уровень постройкой орбитальной станции.

```python
df1 = df.groupby('user_id').agg({'event':'nunique', 'event_datetime': ['min', 'max']}).reset_index()
df1.columns = ['user_id', 'finish', 'min', 'max']
df1['duration'] = df1['max'] - df1['min']
df1['duration'] = df1['duration'].dt.total_seconds()
df1
```

```python
sample_1 = df1[df1['finish'] == 2]['duration']
sample_2 = df1[df1['finish'] == 3]['duration']
results = st.ttest_ind(sample_2, sample_1, equal_var=False)
alpha = 0.05
print('P-value:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Не получилось отвергнуть нулевую гипотезу ')
```

Можно сделать вывод, что время завершения уровня в зависимости от метода прохождения различается.

### Наличие взаимосвязи между стратегией развития у пользователей, прошедших уровень, и количеством построек

    H0: взаимосвязь между стратегией развития и количеством построек отсутствует;
    H1: между стратегией развития и количеством построек взаимосвязь существует.

```python
strategy_2 = df_source_building[df_source_building['finish'] == 2]['building_count'].tolist()
strategy_3 = df_source_building[df_source_building['finish'] == 3]['building_count'].tolist()

alpha = 0.05

results = st.mannwhitneyu(strategy_2, strategy_3, True, 'less')

print('P-value:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя'
    ) 
```
